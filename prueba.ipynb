```python
import pandas as pd

excel_file_path = '/content/sample_data/Supertienda.xlsx'
df = pd.read_excel(excel_file_path)

display(df.head())
```

```python
df.info()
```

```python
selected_columns = ['Fecha del pedido', 'Forma de envío', 'Segmento', 'País/Región', 'Región', 'Subcategoría', 'Venta', 'Ganancia']
df_selected = df[selected_columns]

display(df_selected.head(21))
```

# Task
Analizar los datos del archivo "SUPERSTORE.xls" para identificar qué factores influyen en la ganancia, enfocándose en las columnas 'Fecha de pedido', 'Forma de envío', 'Segmento', 'País/Región', 'Región', 'Subcategorías', 'Ventas' y 'Ganancias'.

## Análisis exploratorio de datos (eda) enfocado en la ganancia

### Subtask:
Examinar la distribución de la ganancia (histograma, estadísticas descriptivas).


**Reasoning**:
Calculate and display descriptive statistics for the 'Ganancia' column and generate a histogram to visualize its distribution.



```python
import matplotlib.pyplot as plt

display(df_selected['Ganancia'].describe())

plt.figure(figsize=(10, 6))
plt.hist(df_selected['Ganancia'], bins=50, edgecolor='black')
plt.title('Distribución de Ganancia')
plt.xlabel('Ganancia')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

## Análisis de la relación entre las características y la ganancia

### Subtask:
Para características categóricas (Forma de envío, Segmento, País/Región, Región, Subcategoría), calcular la ganancia promedio por categoría y visualizar la relación utilizando gráficos de barras o box plots.


**Reasoning**:
Calculate the average profit for each category in the specified categorical columns and store the results.



```python
shipping_mode_profit = df_selected.groupby('Forma de envío')['Ganancia'].mean()
segment_profit = df_selected.groupby('Segmento')['Ganancia'].mean()
country_profit = df_selected.groupby('País/Región')['Ganancia'].mean()
region_profit = df_selected.groupby('Región')['Ganancia'].mean()
subcategory_profit = df_selected.groupby('Subcategoría')['Ganancia'].mean()

display(shipping_mode_profit)
display(segment_profit)
display(country_profit)
display(region_profit)
display(subcategory_profit)
```

**Reasoning**:
Generate bar plots for the average profit of each category in the specified categorical columns.



```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 20))
axes = axes.flatten()

shipping_mode_profit.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Ganancia Promedio por Forma de Envío')
axes[0].set_xlabel('Forma de Envío')
axes[0].set_ylabel('Ganancia Promedio')
axes[0].tick_params(axis='x', rotation=45)

segment_profit.plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title('Ganancia Promedio por Segmento')
axes[1].set_xlabel('Segmento')
axes[1].set_ylabel('Ganancia Promedio')
axes[1].tick_params(axis='x', rotation=45)

country_profit.plot(kind='bar', ax=axes[2], color='salmon')
axes[2].set_title('Ganancia Promedio por País/Región')
axes[2].set_xlabel('País/Región')
axes[2].set_ylabel('Ganancia Promedio')
axes[2].tick_params(axis='x', rotation=90)

region_profit.plot(kind='bar', ax=axes[3], color='gold')
axes[3].set_title('Ganancia Promedio por Región')
axes[3].set_xlabel('Región')
axes[3].set_ylabel('Ganancia Promedio')
axes[3].tick_params(axis='x', rotation=45)

subcategory_profit.plot(kind='bar', ax=axes[4], color='violet')
axes[4].set_title('Ganancia Promedio por Subcategoría')
axes[4].set_xlabel('Subcategoría')
axes[4].set_ylabel('Ganancia Promedio')
axes[4].tick_params(axis='x', rotation=90)

fig.delaxes(axes[5]) # Remove the unused subplot

plt.tight_layout()
plt.show()
```

## Análisis de la relación entre las características y la ganancia

### Subtask:
Para características numéricas (Venta), calcular la correlación entre Venta y Ganancia y visualizar la relación utilizando un gráfico de dispersión.


**Reasoning**:
Calculate the correlation coefficient between 'Venta' and 'Ganancia' and visualize the relationship using a scatter plot.



```python
import matplotlib.pyplot as plt

correlation = df_selected['Venta'].corr(df_selected['Ganancia'])
print(f"Coeficiente de correlación entre Venta y Ganancia: {correlation}")

plt.figure(figsize=(10, 6))
plt.scatter(df_selected['Venta'], df_selected['Ganancia'], alpha=0.5)
plt.title('Relación entre Venta y Ganancia')
plt.xlabel('Venta')
plt.ylabel('Ganancia')
plt.grid(True)
plt.show()
```

## Análisis de la relación entre las características y la ganancia

### Subtask:
Para la fecha (Fecha del pedido), analizar la ganancia a lo largo del tiempo (por mes, trimestre o año) y visualizar las tendencias temporales.


**Reasoning**:
Ensure the 'Fecha del pedido' column is in datetime format, extract year, month, and quarter, calculate average profit by year, month, and quarter, and visualize the trends.



```python
df_selected['Fecha del pedido'] = pd.to_datetime(df_selected['Fecha del pedido'])

df_selected['Año'] = df_selected['Fecha del pedido'].dt.year
df_selected['Mes'] = df_selected['Fecha del pedido'].dt.month
df_selected['Trimestre'] = df_selected['Fecha del pedido'].dt.quarter

profit_by_year = df_selected.groupby('Año')['Ganancia'].mean()
profit_by_month = df_selected.groupby('Mes')['Ganancia'].mean()
profit_by_quarter = df_selected.groupby('Trimestre')['Ganancia'].mean()

import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 15))

profit_by_year.plot(kind='line', ax=axes[0], marker='o', color='blue')
axes[0].set_title('Ganancia Promedio por Año')
axes[0].set_xlabel('Año')
axes[0].set_ylabel('Ganancia Promedio')
axes[0].grid(True)

profit_by_month.plot(kind='line', ax=axes[1], marker='o', color='green')
axes[1].set_title('Ganancia Promedio por Mes')
axes[1].set_xlabel('Mes')
axes[1].set_ylabel('Ganancia Promedio')
axes[1].set_xticks(range(1, 13))
axes[1].grid(True)

profit_by_quarter.plot(kind='line', ax=axes[2], marker='o', color='red')
axes[2].set_title('Ganancia Promedio por Trimestre')
axes[2].set_xlabel('Trimestre')
axes[2].set_ylabel('Ganancia Promedio')
axes[2].set_xticks(range(1, 5))
axes[2].grid(True)

plt.tight_layout()
plt.show()
```
