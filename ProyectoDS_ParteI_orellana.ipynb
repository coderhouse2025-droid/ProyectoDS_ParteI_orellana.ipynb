```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

if 'df' not in locals():
    df = pd.read_excel('/content/sample_data/Supertienda.xlsx')


segmento_sales = df.groupby('Segmento')['Venta'].sum()

plt.figure(figsize=(16, 9))
plt.pie(segmento_sales.values, labels=segmento_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribución de Ventas por Segmento')
plt.axis('equal')
plt.show()
```

PRODUCTOS MAS VENDIDOS POR SUPERTIENDA

```python
top_products = df.groupby('Nombre del producto')['Cantidad'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(16, 9))
sns.barplot(x=top_products.index, y=top_products.values, palette='viridis', hue=top_products.index, legend=False)
plt.title('TOP 10 PRODUCTOS CON MAYOR VOLUNEN DE VENTAS')
plt.xlabel('NOMBRE DEL PRODUCTO')
plt.ylabel('VENTAS')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

VENTAS REGIONALES DE SUPERTIENDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.ticker as mtick

regional_country_sales = df.groupby(['Región', 'País/Región'])['Venta'].sum().reset_index()

top_countries_per_region = regional_country_sales.loc[regional_country_sales.groupby('Región')['Venta'].nlargest(3).index.get_level_values(1)]

top_countries_per_region_df = pd.DataFrame(top_countries_per_region)

pivot_df = top_countries_per_region_df.pivot_table(index='Región', columns='País/Región', values='Venta', fill_value=0)

top_3_sales_per_region = {}
for region in pivot_df.index:
    top_3_countries_sales = pivot_df.loc[region].sort_values(ascending=False).head(3)
    top_3_sales_per_region[region] = top_3_countries_sales

plot_data = pd.DataFrame(top_3_sales_per_region).T

plot_data = plot_data.apply(lambda x: x.sort_values(ascending=False), axis=1)

plt.figure(figsize=(16, 9))

ax = plot_data.plot(kind='bar', stacked=True, figsize=(16, 9), colormap='viridis')

plt.title('VENTAS POR REGION')
plt.xlabel('REGION')
plt.ylabel('VENTAS (USD)')
plt.xticks(rotation=0)

formatter = mtick.FormatStrFormatter('$%1.0f')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

plt.tight_layout()
plt.show()
```

VENTAS DEL PERIODO DE SUPERTIENDA

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

df['Año del pedido'] = df['Fecha del pedido'].dt.year
ventas_anuales = df.groupby('Año del pedido')['Venta'].sum()

plt.figure(figsize=(16, 9))
plt.plot(ventas_anuales.index, ventas_anuales.values, marker='o', linestyle='-')
plt.title('EVOLUCION DE LAS VENTAS')
plt.xlabel('EJERCICIO')
plt.ylabel('VENTAS (USD)')
plt.grid(True)

ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
ax.yaxis.set_major_locator(plt.MultipleLocator(2000000))
plt.ylim(0, 13000000)

ax.xaxis.set_major_locator(plt.MultipleLocator(1))
plt.xticks(ventas_anuales.index.astype(int))

for year, sales in ventas_anuales.items():
    ax.annotate(f'${sales:,.0f}', (year, sales), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```
